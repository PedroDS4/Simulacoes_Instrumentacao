```python
import numpy as np
import matplotlib.pyplot as plt


N = 10000  #Número de simulações

R_med = 100 #Ohm
mu_R = 5 # Ohm

V_med = 2 #V
mu_V = 0.05 #V


V = V_med + np.random.normal(0, mu_V, N)
R = R_med + np.random.normal(0, mu_R, N)



#Potência Dissipada no resistor
P = V**2/R


P_med = round( np.mean(P), 4)
mu_P = round(np.std(P),4)

print("Potência média dissipada = ", P_med)
print("Incerteza da potência dissipada = ", mu_P)


print(f'A potência dissipada é dada por P = {P_med} ± {mu_P} W ')

plt.hist(P, bins=100)
plt.xlabel('Potência dissipada (W)')
plt.grid()
plt.ylabel('Ocorrência')
plt.title(f'Histograma da potência dissipada com média $\\overline{{P}}$ = {P_med} e incerteza $\\mu_P$ = {mu_P}')
plt.show()
```
